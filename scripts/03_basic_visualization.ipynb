## Line Plots and the Basics of Visualization

```julia
# create a basic line plots
x = range(0, 10, length = 100)
y1 = sin.(x)
y2 = cos.(x)
plot(x, [y1, y2])
```

```julia
# modify existing plot by adding one more series
y3 = @. sin(x)^2 - 1/2
plot!(x, y3)
```

```julia
# developing graphics by using plot object (same as previous plot)

x = range(0, 10, length=100)
y1 = sin.(x)
y2 = cos.(x)
p = plot(x, [y1 y2])  # p is the plot object

y3 = @. sin(x)^2 - 1/2
plot!(p, x, y3)
```

```julia
# saving plots to file
savefig("figures/my_cruurent_plot.png") # saves the current plot avaialble in the environment
savefig(p, "figures/my_plot_object.png")
```

```julia
# adding attributes to make the visualization better
# adding Latex Strings to make texts more appealing
# we can use the alias `lw` for `linewidth`
# we can use the alias `lc` for `linecolor`
# we can use the alias `mc` for `markercolor`
# we can use the alias `ms` for `markersize`
# we can use the alias `ma` for `markeralpha`


x = range(0, 10, length = 100)
y1 = @. sin(x)
y1 = @. sin(x)
p = plot(x, [y1, y2], 
title="Trigonometric Functions", 
label=[L"sin(x)" L"cos(x)"], 
linewidth=1)

xlims!(0, 2)
xlabel!(L"x")
ylabel!(L"y")
```

## Scatter Plots

```julia
# prepare data
x = range(0, 10, length = 100)
y = @. sin(x)
y_noise = @. sin(x) + 0.1 * randn()

plt = plot(x, y, label="sin(x)")
xlims!(0, 2)
xlabel!(L"x")
# ylabel!("noisy sine wave")
title!("An artificial data")
plot!(plt, x, y_noise, seriestype=:scatter, label="data")
plot!(legend=:bottomleft)
```

## Aligning lots in a grid - creating subplots

```julia
# aligning plots into a grid - creating subplots
x = range(0, 10, length = 100)
y1 = @. exp(-0.1x) * cos(4x)
y2 = @. exp(-0.3x) * cos(4x)
y3 = @. exp(-0.5x) * cos(4x)
plot(x, [y1, y2, y3], layout=(3,1), legend=false)
```

## Bonus: Understanding your data through histogram

```julia
# prepare data
N = 10^3
x = rand(Normal(3, 0.05), N)

h = histogram(x, normalize=:pdf, color=:lightblue)

plot!(h, title="Density Type Histogram of the Data", xlabel="data", ylabel="probability density", legend=false)
```
